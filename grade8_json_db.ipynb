```python
import requests
import json
import firebase_admin
from firebase_admin import credentials, db
# Step 1: Download the key from GitHub
url = "https://raw.githubusercontent.com/EliasDAraneta/Quarter3-Codes/refs/heads/main/Students.JSON"
response = requests.get(url)
# Step 2: Save it locally in Colab
with open("/content/grade8-json-db-29f1f-firebase-adminsdk-fbsvc-861d28478a.json", "wb") as f:
  f.write(response.content)
# Step 3: Initialize Firebase
# Check if the default Firebase app is already initialized
if not firebase_admin._apps:
  cred = credentials.Certificate("/content/grade8-json-db-29f1f-firebase-adminsdk-fbsvc-861d28478a.json")
  firebase_admin.initialize_app(cred, {'databaseURL': 'https://grade8-json-db-29f1f-default-rtdb.asia-southeast1.firebasedatabase.app/'})
  print("Connected to Firebase!")
else:
  print("Firebase app already initialized.")
#make sure the url is the raw file from the github
```
